```python
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt

# Load pre-trained VGG16 model
model = models.vgg16(pretrained=True)
model.eval()

# Define loss function
criterion = nn.CrossEntropyLoss()

# Load and preprocess the input image
image = Image.open('dog.jpg')
preprocess = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
input_tensor = preprocess(image).unsqueeze(0)

# Set the requires_grad flag to generate gradients
input_tensor.requires_grad = True

# Perform forward pass
output = model(input_tensor)
original_prediction = torch.argmax(output)

target_label = 10

# Calculate gradients using backpropagation
model.zero_grad()
loss = criterion(output, torch.tensor([target_label]))
loss.backward()

# Generate adversarial example using FGSM
epsilon = 0.3  # Perturbation magnitude
sign_grad = torch.sign(input_tensor.grad.data)
adversarial_image = input_tensor + epsilon * sign_grad

# Perform forward pass on the adversarial example
adversarial_output = model(adversarial_image)
adversarial_prediction = torch.argmax(adversarial_output)

# Print the original and adversarial predictions
print("Original Prediction:", original_prediction.item())
print("Adversarial Prediction:", adversarial_prediction.item())

# Visualize the original and adversarial images
fig, axs = plt.subplots(1, 2)
axs[0].imshow(image)
axs[0].set_title('Original Image')
axs[1].imshow(adversarial_image.squeeze().detach().numpy().transpose((1, 2, 0)))
axs[1].set_title('Adversarial Image')
plt.show()
```
